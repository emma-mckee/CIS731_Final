```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType
from pyspark.sql.types import StructField
from pyspark.sql.types import StringType

spark = SparkSession.builder.appName("AccidentDataProcessing").getOrCreate()

common_schema = StructType([
    StructField('ACCIDENT_KEY', StringType(), True),
    StructField('AGENCY_ORI', StringType(), True),
    StructField('LOCAL_CASE_ID', StringType(), True),
    StructField('COUNTY_NBR', StringType(), True),
    StructField('CITY_NBR', StringType(), True),
    StructField('OFFICER_MILEPOST', StringType(), True),
    StructField('ON_ROAD_KDOT_PFX_DIR', StringType(), True),
    StructField('ON_ROAD_KDOT_NAME', StringType(), True),
    StructField('ON_ROAD_KDOT_TYPE', StringType(), True),
    StructField('ON_ROAD_KDOT_SFX_DIR', StringType(), True),
    StructField('AT_ROAD_KDOT_PFX_DIR', StringType(), True),
    StructField('AT_ROAD_KDOT_NAME', StringType(), True),
    StructField('AT_ROAD_KDOT_TYPE', StringType(), True),
    StructField('AT_ROAD_KDOT_SFX_DIR', StringType(), True),
    StructField('AT_ROAD_KDOT_DISTANCE', StringType(), True),
    StructField('AT_ROAD_KDOT_DIST_UOM', StringType(), True),
    StructField('AT_ROAD_KDOT_DIRECTION', StringType(), True),
    StructField('DOT_LATITUDE', StringType(), True),
    StructField('DOT_LONGITUDE', StringType(), True),
    StructField('NONSTATE_FUNCTION_CLASS', StringType(), True),
    StructField('NONSTATE_REF_POST', StringType(), True),
    StructField('BLOCK_NBR', StringType(), True),
    StructField('ON_ROAD_SPEED_LIMIT', StringType(), True),
    StructField('ON_ROAD_SURFACE_TYPE', StringType(), True),
    StructField('ROAD_SURFTYPE_OTHER', StringType(), True),
    StructField('ON_ROAD_SURFACE_COND', StringType(), True),
    StructField('ROAD_SURFCOND_OTHER', StringType(), True),
    StructField('ON_ROAD_SURFACE_CHAR', StringType(), True),
    StructField('ROAD_SURFCHAR_OTHER', StringType(), True),
    StructField('ON_ROAD_CONST_MAINT', StringType(), True),
    StructField('ON_ROAD_NBR_OF_LANES', StringType(), True),
    StructField('ROAD_NBRLANES_OTHER', StringType(), True),
    StructField('AT_ROAD_SPEED_LIMIT', StringType(), True),
    StructField('AT_ROAD_SURFACE_TYPE', StringType(), True),
    StructField('AT_ROAD_SURFACE_COND', StringType(), True),
    StructField('AT_ROAD_SURFACE_CHAR', StringType(), True),
    StructField('AT_ROAD_CONST_MAINT', StringType(), True),
    StructField('AT_ROAD_NBR_OF_LANES', StringType(), True),
    StructField('DATE_OF_ACCIDENT', StringType(), True),
    StructField('TIME_OF_ACCIDENT', StringType(), True),
    StructField('DAY_OF_ACCIDENT', StringType(), True),
    StructField('DATE_NOTIFIED', StringType(), True),
    StructField('TIME_NOTIFIED', StringType(), True),
    StructField('DAY_NOTIFIED', StringType(), True),
    StructField('DATE_ARRIVED', StringType(), True),
    StructField('TIME_ARRIVED', StringType(), True),
    StructField('DAY_ARRIVED', StringType(), True),
    StructField('ACCIDENT_CLASS_FHE', StringType(), True),
    StructField('ACCIDENT_CLASS_MHE', StringType(), True),
    StructField('ACC_CLASS_OTHER_OBJECT', StringType(), True),
    StructField('COLLISION_W_OTHER_VEH_FHE', StringType(), True),
    StructField('COLLISION_W_OTHER_VEH_MHE', StringType(), True),
    StructField('COLLISION_W_VEH_OTHER', StringType(), True),
    StructField('FIXED_OBJECT_TYPE_FHE', StringType(), True),
    StructField('FIXED_OBJECT_TYPE_MHE', StringType(), True),
    StructField('FIXED_OBJECT_OTHER', StringType(), True),
    StructField('ACCIDENT_LOCATION', StringType(), True),
    StructField('ACCIDENT_LOCATION_OTHER', StringType(), True),
    StructField('INTERSECTION_TYPE', StringType(), True),
    StructField('ROUNDABOUT_NBR', StringType(), True),
    StructField('LIGHT_CONDITIONS', StringType(), True),
    StructField('WEATHER_CONDITIONS', StringType(), True),
    StructField('WEATHER_COND_OTHER', StringType(), True),
    StructField('ANIMAL_TYPE', StringType(), True),
    StructField('KDOT_WORK_ZONE_FLAG', StringType(), True),
    StructField('WORK_ZONE_LOCATION', StringType(), True),
    StructField('WORK_ZONE_CATEGORY', StringType(), True),
    StructField('WORK_ZONE_CAT_OTHER', StringType(), True),
    StructField('OBJECT_DAMAGED1', StringType(), True),
    StructField('KDOT_PROP_DAMAGE_FLAG1', StringType(), True),
    StructField('OBJECT_DAMAGED2', StringType(), True),
    StructField('KDOT_PROP_DAMAGE_FLAG2', StringType(), True),
    StructField('SPECIAL_EVENT', StringType(), True),
    StructField('SPECIAL_DATA', StringType(), True),
    StructField('SPECIAL_JURISDICTION', StringType(), True),
    StructField('SPECIAL_JURIS_OTHER', StringType(), True),
    StructField('ON_ROAD_PREFIX_DIRECTION', StringType(), True),
    StructField('ON_ROAD_NAME', StringType(), True),
    StructField('ON_ROAD_TYPE', StringType(), True),
    StructField('ON_ROAD_SUFFIX_DIRECTION', StringType(), True),
    StructField('ON_ROAD_ALTERNATE_NAME', StringType(), True),
    StructField('AT_ROAD_PREFIX_DIRECTION', StringType(), True),
    StructField('AT_ROAD_NAME', StringType(), True),
    StructField('AT_ROAD_TYPE', StringType(), True),
    StructField('AT_ROAD_SUFFIX_DIRECTION', StringType(), True),
    StructField('AT_ROAD_DISTANCE', StringType(), True),
    StructField('AT_ROAD_DISTANCE_UOM', StringType(), True),
    StructField('AT_ROAD_DIRECTION', StringType(), True),
    StructField('LATITUDE', StringType(), True),
    StructField('LONGITUDE', StringType(), True),
    StructField('S_REF_DISTANCE', StringType(), True),
    StructField('S_REF_ROAD', StringType(), True),
    StructField('W_REF_DISTANCE', StringType(), True),
    StructField('W_REF_ROAD', StringType(), True),
    StructField('OFFICER', StringType(), True),
    StructField('OFFICER_BADGE_ID', StringType(), True),
    StructField('REPORTING_SEVERITY', StringType(), True),
    StructField('AMENDED_REPORT_FLAG', StringType(), True),
    StructField('DUI_FLAG', StringType(), True),
    StructField('PRIVATE_PROPERTY_FLAG', StringType(), True),
    StructField('UAB', StringType(), True)
    # Include all the fields for ACCIDENTS.txt here
])


years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

combined_df = None

for year in years:
    file_path = f"ACCIDENTS_{year}.txt"  # File is in the current directory
    accident = spark.read.csv(file_path, schema=common_schema, header=True, sep="|")

    # If combined_df is not yet initialized, assign the first year's DataFrame to it
    if combined_df is None:
        combined_df = accident
    else:
        # Otherwise, union the new DataFrame with the existing combined one
        combined_df = combined_df.union(accident)

# Show combined DataFrame schema and some data to confirm
combined_df.printSchema()
combined_df.show()
```

```python
columns_to_drop = [
    'AGENCY_ORI',
    'LOCAL_CASE_ID',
    'OFFICER_MILEPOST',
    'ON_ROAD_KDOT_PFX_DIR',
    'ON_ROAD_KDOT_SFX_DIR',
    'AT_ROAD_KDOT_PFX_DIR',
    'AT_ROAD_KDOT_SFX_DIR',
    'WEATHER_COND_OTHER',
    'WEATHER_COND_OTHER',
    'WORK_ZONE_LOCATION',
    'WORK_ZONE_CATEGORY',
    'WORK_ZONE_CAT_OTHER',
    'OBJECT_DAMAGED1',
    'KDOT_PROP_DAMAGE_FLAG1',
    'OBJECT_DAMAGED2',
    'KDOT_PROP_DAMAGE_FLAG2',
    'SPECIAL_EVENT',
    'SPECIAL_DATA',
    'SPECIAL_JURISDICTION',
    'SPECIAL_JURIS_OTHER',
    'ON_ROAD_PREFIX_DIRECTION',
    'ON_ROAD_NAME',
    'ON_ROAD_TYPE',
    'ON_ROAD_SUFFIX_DIRECTION',
    'ON_ROAD_ALTERNATE_NAME',
    'AT_ROAD_PREFIX_DIRECTION',
    'AT_ROAD_NAME',
    'AT_ROAD_TYPE',
    'AT_ROAD_SUFFIX_DIRECTION',
    'AT_ROAD_DISTANCE',
    'AT_ROAD_DISTANCE_UOM',
    'AT_ROAD_DIRECTION',
    'LATITUDE',
    'LONGITUDE',
    'S_REF_DISTANCE',
    'S_REF_ROAD',
    'W_REF_DISTANCE',
    'W_REF_ROAD',
    'OFFICER',
    'OFFICER_BADGE_ID',
    'AMENDED_REPORT_FLAG',
    'PRIVATE_PROPERTY_FLAG',
    'NONSTATE_REF_POST',
    'BLOCK_NBR',
    'ROAD_SURFTYPE_OTHER',
    'ROAD_SURFCOND_OTHER',
    'ROAD_SURFCHAR_OTHER',
    'ROAD_NBRLANES_OTHER',
    'DAY_ARRIVED',
    'TIME_ARRIVED',
    'DATE_ARRIVED',
    'DAY_NOTIFIED',
    'TIME_NOTIFIED',
    'DATE_NOTIFIED',
    'ACC_CLASS_OTHER_OBJECT',
    'COLLISION_W_VEH_OTHER',
    'FIXED_OBJECT_OTHER',
    'ACCIDENT_LOCATION_OTHER',
    'ROUNDABOUT_NBR',
]

# Drop the specified columns from the DataFrame
accident_df = combined_df.drop(*columns_to_drop)

# Show the DataFrame after dropping columns
accident_df.show()
```

```python


schema_accident_summary_2023 = StructType([
    StructField('ACCIDENT_KEY', StringType(), True),
    StructField('NBR_OF_TRAFFIC_UNITS', StringType(), True),
    StructField('NBR_OF_VEHICLES', StringType(), True),
    StructField('NBR_OF_OCCUPANTS', StringType(), True),
    StructField('NBR_OF_PEDESTRIANS', StringType(), True),
    StructField('NBR_OF_FATALITIES', StringType(), True),
    StructField('NBR_OF_DISABLED', StringType(), True),
    StructField('NBR_OF_NON_INCAPACITATING', StringType(), True),
    StructField('NBR_POSSIBLE', StringType(), True),
    StructField('ALCOHOL_INVOLVEMENT', StringType(), True),
    StructField('DRUG_INVOLVEMENT', StringType(), True),
    StructField('ACCIDENT_SEVERITY', StringType(), True)
])

years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
combined_accident_summary_df = None


for year in years:
    file_path = f"ACCIDENT_SUMMARY_{year}.txt"  # File is in the current directory
    accident_summary = spark.read.csv(file_path, schema=schema_accident_summary_2023, header=True, sep="|")

    # If combined_accident_summary_df is not yet initialized, assign the first year's DataFrame to it
    if combined_accident_summary_df is None:
        combined_accident_summary_df = accident_summary
    else:
        # Otherwise, union the new DataFrame with the existing combined one
        combined_accident_summary_df = combined_accident_summary_df.union(accident_summary)

# Show combined DataFrame schema and some data to confirm
combined_accident_summary_df.printSchema()
combined_accident_summary_df.show()
```

```python
schema_vehicle = StructType([
    StructField('ACCIDENT_KEY', StringType(), True),
    StructField('TRAFFIC_UNIT_NBR', StringType(), True),
    StructField('DAMAGE_AREA_NBR', StringType(), True),
    StructField('DAMAGE_AREA_CODE', StringType(), True),
    #StructField('DAMAGE_AREA_OTHER', StringType(), True)
])

years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
combined_vehicle_damage_df = None

for year in years:
    file_path = f"VEHICLE_DAMAGE_AREA_{year}.txt"  # File is in the current directory
    vehicle_damage = spark.read.csv(file_path, schema=schema_vehicle, header=True, sep="|")

    # If combined_vehicle_damage_df is not yet initialized, assign the first year's DataFrame to it
    if combined_vehicle_damage_df is None:
        combined_vehicle_damage_df = vehicle_damage
    else:
        # Otherwise, union the new DataFrame with the existing combined one
        combined_vehicle_damage_df = combined_vehicle_damage_df.unionByName(vehicle_damage)

# Show combined DataFrame schema and some data to confirm
combined_vehicle_damage_df.printSchema()
combined_vehicle_damage_df.show()
```

```python
# Joining the DataFrames on 'ACCIDENT_KEY'
df_combined = accident_df.join(combined_accident_summary_df, 'ACCIDENT_KEY', 'inner') \
                         .join(combined_vehicle_damage_df, 'ACCIDENT_KEY', 'inner')

# Show the combined DataFrame
df_combined.show()
```
